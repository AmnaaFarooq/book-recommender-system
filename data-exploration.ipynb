```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books=pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax= plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)
plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
# put 1 in the books where descript missing
import numpy as np
books["missing_description"]=np.where(books["description"].isna(),1,0)
books["age_of_book"]=2026-books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix= books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matrix,annot=True,fmt =".2f",cmap="coolwarm",cbar_kws={"label":"Spearman Correlation"})
plt.show()

```

```python
#fetch books with no missing description
book_missing= books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
book_missing
```

```python
# sort categories at top the books most famous
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing
```

```python
# finding out  the useless descriptions
book_missing["words_in_description"]= book_missing["description"].str.split().str.len()
# counting the number of words of description
```

```python
book_missing
```

```python
# cut the description books that are short by looking at the data
# dataframe shows us that books upto 250 have long descriptions , we assue they are #correct.
book_missing.loc[book_missing["words_in_description"].between(1,4),"description"]

# useless descriptions
```

```python
book_missing.loc[book_missing["words_in_description"].between(4,15),"description"]

# these are also short descriptions
```

```python
book_missing.loc[book_missing["words_in_description"].between(15,24),"description"]
# somehow useful but not much
```

```python
book_missing.loc[book_missing["words_in_description"].between(25,34),"description"]
# useful
```

```python
# a new dataframe that will have good enough descriptions
book_missing_25_words=book_missing[book_missing["words_in_description"] >=25]
```

```python
book_missing_25_words
# now these are the filtered books
```

```python
# final clearing steps
# combine subtitle and title so that it gives us info,ignore where subtitle is missing , don't combine them
# Ensure everything is a string and replace 'nan' strings with empty space

combined = (
    book_missing_25_words[["title", "subtitle"]]
    .fillna("")
    .astype(str)
    .agg(": ".join, axis=1)
    .str.rstrip(": ") # Clean up the separator if subtitle was empty
)

book_missing_25_words["title_and_subtitle"] = combined

```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"]=book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words

# now we can apply NLP to it
```

```python
# drop columns and clean data
(book_missing_25_words
.drop(["subtitle","missing_description","age_of_book","words_in_description"], axis=1)
.to_csv("books_cleaned.csv",index=False)
 )
# now we have clean data  set finallyyy
```
